```python
import matplotlib.pyplot as plt
import numpy as np

# Constants
accounts = 11
total_tokens = 1_00_50_00_05_00_000  # Correct formatting for Python long integers
total_time = 174  # in seconds

# Calculate average latency and throughput
average_latency = total_time / accounts  # Time per account
throughput = total_tokens / total_time  # Tokens per second

# Data for plotting
accounts_labels = [f"Account {i+1}" for i in range(accounts)]
tokens_per_account = [total_tokens / accounts] * accounts  # Assuming evenly distributed

# Creating the plot
fig, ax1 = plt.subplots()

# Bar plot for tokens transferred
ax1.bar(accounts_labels, tokens_per_account, color='b')
ax1.set_xlabel('Accounts')
ax1.set_ylabel('Tokens transferred', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Secondary axis for latency or throughput
ax2 = ax1.twinx()
ax2.plot(accounts_labels, [average_latency] * accounts, 'r-')  # Average latency as a line
ax2.set_ylabel('Average latency (s)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

plt.xticks(rotation=45)
plt.title('Token Transfers per Account with Latency')
plt.show()

# Print calculated averages
print(f"Average Latency: {average_latency} seconds per account")
print(f"Throughput: {throughput} tokens per second")
```
